Data contains missing values in Item_Weight and Outlet_Size. Let's try to fill missing values in it

```python
train_data.loc[:,['Item_Identifier','Item_Weight']][train_data.Item_Identifier=='FDP10']
```

A close look reveals Item_weight can be identified from Unique Item_Identifier. As seen for FDP10 value

```python
item_weight ={}
for i, j in zip(train_data.Item_Identifier,train_data.Item_Weight):
    if np.isnan(j):
        continue
    else:
        item_weight[i]=j

new_item_weight=[]
for i, j in zip(train_data.Item_Identifier,train_data.Item_Weight):
    if np.isnan(j):
        if i in item_weight:
            new_item_weight.append(item_weight[i])
        else:
            new_item_weight.append(j)
    else:
        new_item_weight.append(j)
train_data.Item_Weight=new_item_weight

new_item_weight=[]
for i, j in zip(test_data.Item_Identifier,test_data.Item_Weight):
    if np.isnan(j):
        if i in item_weight:
            new_item_weight.append(item_weight[i])
        else:
            new_item_weight.append(j)
    else:
        new_item_weight.append(j)
test_data.Item_Weight=new_item_weight
```

There are still some missing values in both train and test_data. We will simply remove the row from training data but fill missing value in test_data by median weight of train data

```python
train_data= train_data[train_data.Item_Weight.notnull()]
```

```python
train_data.shape
```

```python
median= train_data.Item_Weight.median()
print(median)
```

```python
test_data.Item_Weight.fillna(median, inplace=True)
```

So we have removed missing values in Item Weight column. Next move towards outlet size. Let's see if outlet size is related to Outlet Identifier, Outlet location type and outlet type. we will use cross table feature

```python
ct = pd.crosstab(train_data['Outlet_Size'], train_data['Outlet_Type'])

# Display the table as an image
ct
```

As you can see all 'grocery store' has small outlet and all supermarket type 2 and type 3 outlet are medium size

```python
ct = pd.crosstab(train_data['Outlet_Establishment_Year'], train_data['Outlet_Size'])

# Display the table as an image
ct
```

All Tier 2 location type are small outlet size. And Tier 1 location do not have high outlet size. so we can fill 'small' outlet size as it is more in number. Similary we can use year variable to impute values of outlet size

```python
outlet_size =[]
for loc_type,size,outlet_type,year in zip(train_data.Outlet_Location_Type,train_data.Outlet_Size,train_data.Outlet_Type,train_data.Outlet_Establishment_Year):
    if type(size)==float:
        if outlet_type=='Supermarket Type3' or outlet_type=='Supermarket Type2':
            outlet_size.append('Medium')
            continue
        elif outlet_type=='Grocery Store':
            outlet_size.append('Small')
            continue
        elif year==1987:
            outlet_size.append('High')
            continue
        elif year==1997 or year==2004:
            outlet_size.append('Small')
            continue
        elif year==1999 or year==2009:
            outlet_size.append('Small')
            continue
        elif loc_type=='Tier 2':
            outlet_size.append('Small')
            continue 
        elif loc_type=='Tier 3':
            outlet_size.append('Medium')
            continue 
        elif loc_type=='Tier 1':
            outlet_size.append('Small')
            continue
    else:
        outlet_size.append(size)
train_data.Outlet_Size= outlet_size
```

```python
outlet_size =[]
for loc_type,size,outlet_type,year in zip(test_data.Outlet_Location_Type,test_data.Outlet_Size,test_data.Outlet_Type,test_data.Outlet_Establishment_Year):
    if type(size)==float:
        if outlet_type=='Supermarket Type3' or outlet_type=='Supermarket Type2':
            outlet_size.append('Medium')
            continue
        elif outlet_type=='Grocery Store':
            outlet_size.append('Small')
            continue
        elif year==1987:
            outlet_size.append('High')
            continue
        elif year==1997 or year==2004:
            outlet_size.append('Small')
            continue
        elif year==1999 or year==2009:
            outlet_size.append('Small')
            continue
        elif loc_type=='Tier 2':
            outlet_size.append('Small')
            continue 
        elif loc_type=='Tier 3':
            outlet_size.append('Medium')
            continue 
        elif loc_type=='Tier 1':
            outlet_size.append('Small')
            continue
    else:
        outlet_size.append(size)
test_data.Outlet_Size= outlet_size
```

Let's Label Encode our Item Identifier

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
le.fit((train_data.Item_Identifier).append(test_data.Item_Identifier))
train_data.Item_Identifier=le.transform(train_data.Item_Identifier).astype(int)
test_data.Item_Identifier=le.transform(test_data.Item_Identifier).astype(int)
```

```python
train_data= pd.get_dummies(train_data)
test_data= pd.get_dummies(test_data)
```
